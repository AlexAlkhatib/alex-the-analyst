# Automatic File Sorter

* utilize Python to automatically sort files in a directory
* includes:
    * importing necessary libraries
        * [`os` module](https://www.w3schools.com/python/module_os.asp?ref=escape.tech) to interact with operating system
        *  [`shutil` module](https://docs.python.org/3/library/shutil.html) contains high-level operations on files and collections of files
    * setting up the path
    * creating sample.csv file, to test the sorter via Kaggle
    * creating the folders:
        * csv files, image files, text files
        * using [for loops](https://www.w3schools.com/python/python_for_loops.asp)
    * sorting the files
        * using [Python conditions](https://www.w3schools.com/python/python_conditions.asp)  

## Import Libraries

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os, shutil
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```

## Set up path

* location of the files you want to sort
* location of the new folders you will create
* you can use local path for your local files (*e.g."C:\Users\User\Documents"*)
* be consistent with the use / or \

```python
path = r"/kaggle/working/" # path assigned via Kaggle Output
```

## Create sample.csv file

* In order to test the sorter here in Kaggle, create a .csv file to sort into the csv files folder.
* No need to peform this step if you're running the sorter locally, and go ahead create the folders.

```python
import csv

# Data to write to the CSV file
data = [
    ['Name', 'Age', 'City'],
    ['Alice', '30', 'New York'],
    ['Bob', '25', 'London'],
    ['Charlie', '35', 'Paris']
]

# Open the CSV file in write mode
with open('sample.csv', 'w', newline='') as file:
    # Create a csv.writer object
    writer = csv.writer(file)

    # Write the data to the CSV file
    writer.writerows(data)

print("CSV file 'sample.csv' created successfully.")
```

## Create the folders
* generates three folder categories: csv files, image files, text files
* feel free to add more folder categories if needed

```python
file_name = os.listdir(path) #lists the files in path 

folder_names = ['csv files', 'image files', 'text files'] #creates folder list

for loop in range(0,3): #checks if that path exists in a loop from position 0 to 3
    if not os.path.exists(path + folder_names[loop]): #if it does not exist, create the folders
        print(path + folder_names[loop])
        os.makedirs(path + folder_names[loop])#prints the folder created
```

## Sorting the Files
* sorts the files using Python conditions
* file extension included: .csv, .jpg, .txt
* feel free to add additional elif statements, with the respective path, for other file extensions (*e.g. .pdf, .xlsx*)  

```python
for file in file_name:
    if ".csv" in file and not os.path.exists(path + "csv files/" + file):
        shutil.move(path + file, path + "csv files/" + file)
    elif ".jpg" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
    elif ".txt" in file and not os.path.exists(path + "text files/" + file):
        shutil.move(path + file, path + "text files/" + file)
        
#will loop in files under file_name path
#checks if the csv file,  jpg file, and txt file are present
#if not present, move the file to the designated file folder
```

**After Sorting**
